### Task 4

```python
with open("/home/jovyan/shared-data/bigdata20/emojis_sentiment.json", "r") as f:
    emojis_sentiment_dict = json.loads(f.read())
```

```python
def task_4(df: "pyspark.sql.dataframe.DataFrame",
           F: "pyspark.sql.functions",
           T: "pyspark.sql.types",
           emojis_data: dict,
           broadcast_func: "spark.sparkContext.broadcast") -> 'Tuple["pyspark.sql.dataframe.DataFrame"]':
    
    # You are able to modify any code inside this function.
    # Only 'emoji' package import is allowed for this task.
    
    import emoji
    
    broadcasted_variable = broadcast_func(emojis_data)
    emoji_keys = list(broadcasted_variable.value.keys())
    
    @F.udf(returnType=T.StringType())
    def udf(col):
        return broadcasted_variable.value[col]

    @F.udf(returnType=T.ArrayType(T.StringType()))
    def get_emojis(text):
        if text is None:
            return []
        emojis = [e for e in text if e in emoji.UNICODE_EMOJI['en']]
        return emojis

    emojis_df = (
        df
        .select("text")
        .withColumn("emoji", F.explode(get_emojis(F.col("text"))))
        .filter(F.col("emoji").isin(emoji_keys))
        .groupby("emoji")
        .agg(F.count("emoji").name("count"))
        .withColumn("sentiment", udf("emoji"))
    )
    
    positive_emojis = (
        emojis_df
        .filter(F.col("sentiment") == "positive")
        .select("emoji", "count")
        .orderBy(F.col("count").desc(),F.col("emoji").asc())
    )
    
    neutral_emojis = (
        emojis_df
        .filter(F.col("sentiment") == "neutral")
        .select("emoji", "count")
        .orderBy(F.col("count").desc(),F.col("emoji").asc())
    )
    
    negative_emojis = (
        emojis_df
        .filter(F.col("sentiment") == "negative")
        .select("emoji", "count")
        .orderBy(F.col("count").desc(),F.col("emoji").asc())
    )

    return (positive_emojis, neutral_emojis, negative_emojis)
```

```python
positive_emojis, neutral_emojis, negative_emojis = task_4(
    df=followers_posts_df,
    F=F,
    T=T,
    emojis_data=emojis_sentiment_dict,
    broadcast_func=spark.sparkContext.broadcast
)
```

### Task 5

```python
def task_5(df: "pyspark.sql.dataframe.DataFrame",
           F: "pyspark.sql.functions",
           W: "pyspark.sql.window.Window",
           top_n_likers: int) -> "pyspark.sql.dataframe.DataFrame":
    
    w_func = W.partitionBy("ownerId").orderBy(F.col("count").desc())
    
    modified_df = (
        df
        .select("ownerId", "likerId")
        .filter(F.col("ownerId")!=F.col("likerId"))
        .groupby("ownerId", "likerId")
        .agg(F.count("likerId").name("count"))
        .withColumn("top_10", F.row_number().over(w_func))
        .filter(F.col("top_10") <= top_n_likers)
        .orderBy(F.col("ownerId").asc(),F.col("count").desc(),F.col("likerId").asc())
    )
    
    modified_df = modified_df.drop("top_10")
    
    return modified_df
```

```python
result_5 = task_5(
    df=followers_posts_likes_df,
    F=F,
    W=Window,
    top_n_likers=10
)
result_5.show(15)
```

### Task 6

```python
def task_6(df: "pyspark.sql.dataframe.DataFrame",
           F: "pyspark.sql.functions",
           W: "pyspark.sql.window.Window") -> "pyspark.sql.dataframe.DataFrame":
    
    w_func = W.partitionBy("likerId").orderBy(F.col("count").desc())
    
    top_fan = (
        df
        .select("likerId", "ownerId")
        .filter(F.col("ownerId")!=F.col("likerId"))
        .groupby("likerId", "ownerId")
        .agg(F.count("ownerId").name("count"))
        .withColumn("user_fans", F.rank().over(w_func))
        .filter(F.col("user_fans") == 1)
        .select("ownerId", "likerId", "count")
    )
    
    modified_df = (
        top_fan.alias("user_a")
        .join(
            top_fan.alias("user_b"),
            [F.col("user_a.ownerId") == F.col("user_b.likerId"),
             F.col("user_b.ownerId") == F.col("user_a.likerId"),
             F.col("user_a.ownerId") < F.col("user_b.ownerId")]
        )
        .withColumn("mutual_likes", F.col("user_a.count") + F.col("user_b.count"))
        .select(F.col("user_a.ownerId").name("user_a"),
                F.col("user_a.likerId").name("user_b"),
                F.col("user_b.count").name("likes_from_a"),
                F.col("user_a.count").name("likes_from_b"),
                F.col("mutual_likes"))
        .orderBy(F.col("mutual_likes").desc(),F.col("user_a").asc(),F.col("user_b").asc())
    )

    
    return modified_df
```

```python
result_6 = task_6(
    df=followers_posts_likes_df,
    F=F,
    W=Window
)
```
